```python
import numpy as np
import matplotlib.pyplot as plt

# 시뮬레이션 설정
initial_money = 1  # 초기 자본
num_trials = 10000  # 시뮬레이션 반복 횟수
num_steps = 500  # 최대 진행 라운드

# 다양한 배팅 비율 테스트 (0.01 ~ 1.0)
bet_fractions = np.linspace(0.01, 1.0, 20)

# 결과 저장
final_wealths = []

for bet_fraction in bet_fractions:
    wealths = []
    for _ in range(num_trials):
        money = initial_money
        for _ in range(num_steps):
            bet = bet_fraction * money
            if np.random.rand() < 0.51:  # 승리
                money += bet
            else:  # 패배
                money -= bet
            if money <= 0:  # 파산
                money = 0
                break
        wealths.append(money)
    final_wealths.append(np.mean(wealths))  # 평균 최종 자본 저장

# 그래프 출력
plt.figure(figsize=(10, 6))
plt.plot(bet_fractions, final_wealths, marker='o', linestyle='-')
plt.xlabel("Bet Fraction (x%)")
plt.ylabel("Final Average Wealth")
plt.title("Effect of Bet Fraction on Final Wealth")
plt.grid(True)
plt.show()
```
